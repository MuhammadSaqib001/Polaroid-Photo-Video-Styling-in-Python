```python
from tkinter import *
image_gui1= Tk()
```

```python
def print_sapce(x):
    for y in range(0,x):
        my_space=Label(text="")
        my_space.pack()
        
def print_text(x):
    my_text=Label(text=x)
    my_text.pack()
```

```python
def window_1():
    
    global filenam
    filename= filenam.get()
    global clicked
    filetype= clicked.get()

    image_gui1.destroy()
    image_gui2= Tk()
    image_gui2.geometry("800x600")
    image_gui2.maxsize(800,600)
    image_gui2.minsize(800,600)
    image_gui2.title("Polaroid --> DIP Project")

    print_sapce(1)
    my_logo1=PhotoImage(file="Polaroid.png")
    lgo_label1=Label(image=my_logo1)
    lgo_label1.pack()
    
    company_label=Label(text="Polaroid")
    company_label.config(font=('Helvetica bold',27))
    company_label.pack()

    print_text("       Styling is to be perfomed on this "+filetype)

    print_sapce(1)
    my_logo=PhotoImage(file=filename)
    lgo_label=Label(image=my_logo)
    lgo_label.pack()

    styling_options = [
        "Video              ",
        "Image              "
    ]
#     clicked = StringVar()
    global clicked 
    clicked= StringVar()

    clicked.set( "Select the desired styling" )
    drop = OptionMenu( image_gui1 , clicked , *styling_options )
    drop.pack()

    
    image_gui2.mainloop()
```

```python
def intial_window():


    # def callback():
    #     Label(image_gui, text="Hello World!", font=('Century 20 bold')).pack(pady=4)
    # btn=Button(image_gui, text="Press Enter", command= callback)

    # btn.pack(ipadx=10)
    # image_gui.bind('<Return>',lambda event:callback())

    #Display Logo 

    image_gui1.geometry("800x600")
    image_gui1.maxsize(800,600)
    image_gui1.minsize(800,600)
    image_gui1.title("Polaroid --> DIP Project")
    print_sapce(1)
    my_logo=PhotoImage(file="Polaroid.png")
    lgo_label=Label(image=my_logo)
    lgo_label.pack()
    company_label=Label(text="Polaroid")
    company_label.config(font=('Helvetica bold',27))
    company_label.pack()

    #Description

    company_description=Label(text="Polaroid provides different styling mechanisms for images and videos.")
    company_description.config(font=('Helvetica bold',10))

    company_description.pack()

    #
    print_sapce(2)
    options = [
        "Video              ",
        "Image              "
    ]
#     clicked = StringVar()
    global clicked 
    clicked= StringVar()

    clicked.set( "Select file type" )
    drop = OptionMenu( image_gui1 , clicked , *options )
    drop.pack()

    print_sapce(1)

    print_text("Enter Path of the File")

    #input image name
    global filenam
    filenam= Entry(image_gui1, width= 30)
    filenam.focus_set()
    filenam.pack()



    print_sapce(1)
    image_submission = Button(image_gui1,text ="Submit File", command = window_1 )
    image_submission.pack()

    image_gui1.mainloop()
```

```python
intial_window()
```
